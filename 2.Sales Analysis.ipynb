```python
import pandas as pd
```

```python
# Load Customers Data
customers_df = pd.read_excel(r"C:\Projects\Global Super Store\2.Processing\Sales Super Store Dataset.xlsx",sheet_name='Customer')
print(customers_df.head())
```

```python
# Load Global Superstore Data
global_superstore_df = pd.read_excel(r"C:\Projects\Global Super Store\2.Processing\Sales Super Store Dataset.xlsx",sheet_name ='Global-Superstore')
print(global_superstore_df.head())
```

```python
# Load Products Data
product_df = pd.read_excel(r"C:\Projects\Global Super Store\2.Processing\Sales Super Store Dataset.xlsx",sheet_name ='Product')
print(product_df.head())
```

```python
# Load Segments Data
segment_df = pd.read_excel(r"C:\Projects\Global Super Store\2.Processing\Sales Super Store Dataset.xlsx",sheet_name ='Segment')
print(segment_df.head())
```

```python
# Load Shipping Modes Data
ship_mode_df = pd.read_excel(r"C:\Projects\Global Super Store\2.Processing\Sales Super Store Dataset.xlsx",sheet_name ='Ship Mode')
print(ship_mode_df.head())
```

### 1.How many Customers made orders for each year?

```python
# Table: global_superstore_df
# Columns:
#  Order Date: Date value of each transaction 
#  Customer ID: ID value of each customer
total_customers_df = global_superstore_df.copy()
total_customers_df['Year'] = total_customers_df['Order Date'].dt.year

# Calculating total unique customers
result_df = total_customers_df.groupby('Year')['Customer ID'].nunique()
print(result_df)
```

### 2.How many Products sold for each Year?

```python
# Table: global_superstore_df
# Columns:
#  Order Date: Date value of each transaction 
#  Product ID: ID value of each product
total_products_df = global_superstore_df.copy()
total_products_df['Year'] = total_products_df['Order Date'].dt.year

# Calculating total unique products
result_df = total_products_df.groupby('Year')['Product ID'].nunique()
print(result_df)
```

### 3.What is the Total value of Shipping Cost for each year?

```python
# Table: global_superstore_df
# Columns:
#  Order Date: Date value of each transaction 
#  Shipping Cost: Shipping Cost value of each order
total_shipping_cost = global_superstore_df.copy()
total_shipping_cost['Year'] = total_shipping_cost['Order Date'].dt.year
result_df = total_shipping_cost.groupby('Year')['Shipping Cost'].sum()
print(result_df)
```

### 4.What is the Total Sale value of each Year?

```python
# Table: global_superstore_df
# Columns:
#  Order Date: Date value of each transaction 
#  Sales: Sales value of each transaction     
total_sales_df = global_superstore_df.copy()
total_sales_df['Year'] = total_sales_df['Order Date'].dt.year
result_df = total_sales_df.groupby('Year')['Sales'].sum()
print(result_df)
```

### 5.What are Top 10 Most Profitable Products and their Sales?

```python
# Table: global_superstore_df, product_df
# Columns:
#  Product ID: ID value of each transaction
#  Product Name: Product Name value of each transaction
#  Profit: Profit value of each transaction
#  Sales: Sale value of each transaction

# Dropping duplicates from product_df to get right results
products_clean = product_df.copy().drop_duplicates(subset=["Product ID","Product Name"])

df = global_superstore_df.copy().merge(products_clean,how='left',on='Product ID')
result_df = df.groupby(by='Product Name',as_index=False,observed=True)[['Sales','Profit']].sum()
print(result_df.sort_values(by=['Sales','Profit'],ascending=False).head(10))
```

### 6.What is the Total Sale value by Segment and Ship Mode?

```python
# Table: global_superstore_df, segment_df, ship_mode_df
# Columns:
#  Segment: Segment value of each transaction
#  Ship Mode: Ship Mode value of each transaction
#  Sales: Sale value of each transaction
df = global_superstore_df.copy().merge(segment_df.copy(),how='left',on='Segment Code')
df1 = df.merge(ship_mode_df.copy(),how='left',on='Ship Mode Code')
result_df = df1.groupby(by='Segment',as_index=False,observed=True)['Sales'].sum()

perc = ((result_df['Sales'])/(result_df['Sales']).sum())*100
result_df['percent_share'] = perc.round(1)

result1_df = df1.groupby(by='Ship Mode',as_index=False,observed=True)['Sales'].sum()
perc1 = ((result1_df['Sales'])/(result1_df['Sales']).sum())*100
result1_df['percent_share'] = perc1.round(1)

print(result_df)
print(result1_df)
```

### 7.What is the Profit share by Category and Sub-Category?

```python
# Table: global_superstore_df, product_df
# Columns:
#  Product ID: ID value of each transaction
#  Category: Category value of each transaction
#  Sub-Category: Sub-Category value of each transaction
#  Profit: Profit value of each transaction
#  Sales: Sale value of each transaction

# Dropping duplicates from product_df to get right results
products_clean = product_df.copy().drop_duplicates(subset=["Product ID","Category","Sub-Category"])

df = global_superstore_df.copy().merge(products_clean,how='left',on='Product ID')
# Calculate the total profit for each Category
category_totals = df.groupby('Category',observed=True)['Profit'].sum()

# Map the total profit back onto the main DataFrame
df['Category_Total_Profit'] = df['Category'].map(category_totals)

# 1. Aggregate profit as before
result_df = df.groupby(by=['Category','Sub-Category'], as_index=False, observed=True)['Profit'].sum()

# 2. Add the Category_Total_Profit to the aggregated result_df
# This requires recalculating the Category_Total_Profit on the merged df  or performing a final merge/map if the grouped result is too large.
# A simple merge is cleaner here:
result_df = result_df.merge(category_totals.rename('Category_Total_Profit'), on='Category', how='left')

# 3. Perform the hierarchical sort to match the treemap visualization
final_sorted_df = result_df.sort_values(
    by=['Category_Total_Profit', 'Profit'], 
    ascending=[False, False], 
    ignore_index=True
).drop(columns=['Category_Total_Profit']) # Drop the helper column

print(final_sorted_df.head(15))
```

### 8.What are the Top 10 Profitable Products and their Sales for 2014?

```python
# Table: global_superstore_df, product_df
# Columns:
#  Order Date: Date value of each transaction
#  Product ID: Product ID value of each transaction
#  Product Name: Product Name value of each transaction
#  Profit: Profit value of each transaction
#  Sale: Sale value of each transaction

# Dropping duplicates from products_df to get right results
products_clean = product_df.copy().drop_duplicates(subset=["Product ID","Product Name"])

global_superstore_df['Year'] = global_superstore_df['Order Date'].dt.year

df = global_superstore_df.copy().merge(products_clean,how='left',on='Product ID')

result_df = df.groupby(by=['Year','Product Name'],as_index=False,observed=True)[['Profit','Sales']].sum()
print(result_df.sort_values(by=['Year','Profit'],ascending=[False,False]).head(10))
```

```python
# ---- 1. Clean Product Dimension: Keep only ONE Product Name per Product ID ----

# Option 1: Keep the first product name alphabetically (stable)
products_clean = (
    product_df.sort_values("Product Name")      # sort ensures consistent selection
               .drop_duplicates(subset=["Product ID"], keep="last")
               .reset_index(drop=True)
)

# ---- 2. Extract YEAR from Order Date ----
global_superstore_df["Year"] = global_superstore_df["Order Date"].dt.year

# ---- 3. Merge fact + clean dimension ----
df = global_superstore_df.merge(products_clean, on="Product ID", how="left")

# ---- 4. Group exactly like Tableau ----
result_df = (
    df.groupby(["Year", "Product Name"], as_index=False)["Profit"]
      .sum()
      .sort_values(["Year", "Profit"], ascending=[False, False])
)

# ---- 5. Show top results for 2014 ----
print(result_df[result_df["Year"] == 2014].head(15))
```
