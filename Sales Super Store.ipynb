## 1.How many Customers made orders for each year?

```python
# Table: global_superstore_df
# Columns:
#  Order Date: Date value of each transaction 
#  Customer ID: ID value of each customer
total_customers_df = global_superstore_df
total_customers_df['Year'] = total_customers_df['Order Date'].dt.year
total_customers_df.groupby('Year')['Customer ID'].nunique()
```

## 2.How many Products sold for each Year?

```python
# Table: global_superstore_df
# Columns:
#  Year: Year value of each transaction 
#  Product ID: ID value of each product
total_products_df = global_superstore_df
total_products_df = total_products_df.groupby('Year')['Product ID'].nunique()
total_products_df
```

## 3.What is the Total value of Shipping Cost for each year?

```python
# Table: global_superstore_df
# Columns:
#  Year: Year value of each transaction 
#  Shipping Cost: Shipping Cost value of each order
total_shipping_cost = global_superstore_df
total_shipping_cost = total_shipping_cost.groupby('Year')['Shipping Cost'].sum()
total_shipping_cost
```

## 4.What is the Total Sale value of each Year?

```python
# Table: global_superstore_df
# Columns:
#  Year: Year value of each transaction 
#  Sales: Sales value of each transaction     
total_sales_df = global_superstore_df
total_sales_df = total_sales_df.groupby('Year')['Sales'].sum()
total_sales_df
```

## 5.What is the Total Sale value by each Year and Month?

```python
# Table: global_superstore_df
# Columns:
#  Year: Year value of each transaction 
#  Sales: Sales value of each transaction

month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                   'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

total_sales_df = global_superstore_df

total_sales_df['Month'] = total_sales_df['Order Date'].dt.month_name().str[:3]
total_sales_df['Month_Name'] = pd.Categorical(total_sales_df['Month'], categories=month_order, ordered=True)

total_sales_df = total_sales_df.groupby(['Year','Month_Name'],as_index=False)['Sales'].sum()
total_sales_df.sort_values(by=['Year','Month_Name'])
```
